###BPE algo --> most common pair of consecutive bytes of data is replaced with a byte that does not occue in the data (this algo was about compression)

--> But in LLM it is used in a little different way
- BPE ensure that most common words in the vocab are represented as a single token while Rare words are broken down into two or more subwords tokens


```python
import tiktoken
enc = tiktoken.get_encoding("gpt2")

# Example text to encode
text = "Hello, how are you today? <|endoftext|> This is a Byte Pair Encoding example using tiktoken."
print(f"Original Text: {text}")

# Encode the text into a list of token IDs
token_ids = enc.encode(text, allowed_special={"<|endoftext|>"})
print(f"Token IDs: {token_ids}")
print(f"Number of tokens: {len(token_ids)}")

# Decode the token IDs back into text
decoded_text = enc.decode(token_ids)
print(f"Decoded Text: {decoded_text}")
```
